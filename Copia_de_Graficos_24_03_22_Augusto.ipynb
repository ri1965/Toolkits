# Pandas

```python
# importando librerias necesarias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
clientes = pd.DataFrame(dic_clientes)
clientes = clientes.T
clientes
```

```python
pedidos = pd.DataFrame(dic_pedidos)
pedidos = pedidos.T
pedidos
```

```python
# Ejemplo con función merge

clientesPedidos_merge = clientes.merge(pedidos, how = 'left', left_on = clientes.index, right_on = "idCliente" )
clientesPedidos_merge
```

```python
clientesPedidos = pd.DataFrame()
clientesPedidos = pedidos.join(clientes,on='idCliente')
clientesPedidos.index = clientesPedidos.index.astype(int)
clientesPedidos = clientesPedidos.sort_index()
ClientesPedidos  = clientesPedidos[['razon_social','direccion','localidad','cuit','producto','cantidad','idCliente']]
ClientesPedidos
```

```python
clientesPedidos.info()
```

```python
clientesPedidos.cantidad = clientesPedidos.cantidad.astype(int)
```
